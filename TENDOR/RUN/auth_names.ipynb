```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://www.tender247.com/organization-wise-tenders"

# Send a GET request to the webpage
try:
    response = requests.get(url)
    # Raise an HTTPError for bad responses (4xx or 5xx)
    response.raise_for_status()
except requests.exceptions.RequestException as e:
    print(f"Error fetching the page: {e}")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all <a> tags that are descendants of <li> tags.
# Based on your example, the desired links are within <li> elements.
# We can further refine by looking for the specific class if 'li a' is too broad,
# but let's start with 'li a' as it's often sufficient for lists.
# If needed, you could use: soup.select('li a.text-[#5C5C5C].hover:text-[#2775AB]')
# However, class names with special characters like # might require escaping
# or a different selection method. Let's stick to 'li a' first.
authority_elements = soup.select('li a')

authorities_data = []

for element in authority_elements:
    # Get the text content (the name)
    name = element.get_text(strip=True)

    # Get the href attribute (the link)
    link = element.get('href')

    # Check if both name and link are present and the name is not just whitespace
    if name and link:
        # The href might be a relative path (like "/organization/...")
        # You might want to construct the absolute URL if needed.
        # For this task, just the href value is likely sufficient.
        authorities_data.append({'name': name, 'link': link})

# Print the extracted names and links
if authorities_data:
    print(f"Found {len(authorities_data)} potential authority names and links:")
    for item in authorities_data:
        print(f"Name: {item['name']}, Link: {item['link']}")
else:
    print("No authority names and links found. The selector might need adjustment.")
```
